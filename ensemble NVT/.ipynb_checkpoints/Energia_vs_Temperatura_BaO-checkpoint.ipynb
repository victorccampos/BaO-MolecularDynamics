```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

Importando os csv's das simulações

```python
df20 = pd.read_csv("ET_20.csv") 
df100 = pd.read_csv("ET_100.csv") 
df300 = pd.read_csv("ET_300.csv") 
df500 = pd.read_csv("ET_500.csv") 
df700 = pd.read_csv("ET_700.csv") 
df900 = pd.read_csv("ET_900.csv") 
df1100 = pd.read_csv("ET_1100.csv") 
df1300 = pd.read_csv("ET_1300.csv") 
df1500 = pd.read_csv("ET_1500.csv") 
df1700 = pd.read_csv("ET_1700.csv") 
df2000 = pd.read_csv("ET_2000.csv") 
df2300 = pd.read_csv("ET_2300_2.csv") 
df2600 = pd.read_csv("ET_2600.csv")
df2900 = pd.read_csv("ET_2900.csv") 
df3200 = pd.read_csv("ET_3200.csv") 
df3500 = pd.read_csv("ET_3500.csv") 
```

```python
data_frames = [df20, df100, df300, df500, df700, df900, df1100, df1300,df1500, df1700, df2000, df2300, df2600, df2900, df3200, df3500]
```

```python
Temperaturas = [20, 100, 300, 500, 700, 900, 1100, 1300, 1500, 1700, 2000, 2300, 2600, 2900, 3200, 3500]
```

```python
#print(data_frames[0].columns[0])
"""
data_frames[0].columns[1]

o primeiro [] representa o dataframe a ser acessado.
o segundo indica qual coluna do dataframe a ser acessado, na forma de string "column"

nesse caso 
data_frames[0].columns[1] => Acessa o primeiro df (temperatura 20) e a primeira coluna (temperatura)
data_frames[0].columns[0] => Acessa o primeiro df (temperatura 20) e a coluna zero (MD Steps)
"""
data_frames[0].columns[1]
```

```python
energias_medias = []
erros_energias = []


temperaturas_medias = []
erros_temperaturas = []
i = 0


for data_frame in data_frames:
    
    #print(f" Tabela {i+1} - Temperatura :{Temperaturas[i]} \n {data_frame}")
    #print()
    #print(f" Printa o nome da coluna de energia (string): {data_frame.columns[4]}\n")
    
    
    # Acessa a coluna de energias, nos 20 mil últimos passos : data_frame[data_frame.columns[4]][3960:]
    # retorna o desvio padrão e média de Energia :  np.std(data_frame[data_frame.columns[4]][3960:]) ; np.mean(data_frame[data_frame.columns[4]][3960:])
    # retorna o desvio padrão e média de Temperatura :  np.std(data_frame[data_frame.columns[1]][3960:]) ; np.mean(data_frame[data_frame.columns[1]][3960:])
     
    # Preenche os vetores
    
    energias_medias.append(np.mean(data_frame[data_frame.columns[4]][3960:]))
    temperaturas_medias.append(np.mean(data_frame[data_frame.columns[1]][3960:]))

    erros_energias.append(np.std(data_frame[data_frame.columns[4]][3960:]))
    erros_temperaturas.append(np.std(data_frame[data_frame.columns[1]][3960:]))
    
    #print("-"*50)
    i += 1
    
   
```

```python
E_T = []
E_MD = []
```

```python
for index, data_frame in enumerate(data_frames):
    #print(data_frame[data_frame.columns[4]])
    #plt.plot(data_frame[data_frame.columns[4]], data_frame[data_frame.columns[1]])
    
    #E_T.append( [ data_frame[data_frame.columns[4]] , data_frame[data_frame.columns[1]] ] )
    # últimos passos
    E_T.append( [ data_frame[data_frame.columns[4]][3960:]  , data_frame[data_frame.columns[1]][3960:] ] )
    E_MD.append( [ data_frame[data_frame.columns[4]][:]  , data_frame[data_frame.columns[0]][:] ] )
    
    #E_MD.append( [ data_frame[data_frame.columns[4]][3960:]  , data_frame[data_frame.columns[0]][3960:] ] )
    
    #E_MD.append( [ data_frame[data_frame.columns[4]][:]  , data_frame[data_frame.columns[0]][:] ] )
    
    
    #print(index)
```
