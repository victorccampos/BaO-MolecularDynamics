```python
energias = []
erros_energias = []

temperaturas = []
erros_temperaturas = []

volumes = []
erros_volumes = []


i = 0


for data_frame in data_frames:
    cut = -1000 # pega os ultimos 500 mil passos de cada um
    # energia = 4 / volume = 6 / pressao = 5 / temperatura = 1
    
    energias.append(np.mean(data_frame[data_frame.columns[4]][cut:]))
    temperaturas.append(np.mean(data_frame[data_frame.columns[1]][cut:]))
    volumes.append(np.mean(data_frame[data_frame.columns[6]][cut:]))
    

    erros_energias.append(np.std(data_frame[data_frame.columns[4]][cut:]))
    erros_temperaturas.append(np.std(data_frame[data_frame.columns[1]][cut:]))
```

`Energia versus Temperatura`

```python
plt.scatter(temperaturas, energias, label=r'$E$ x $T$')
plt.xticks(ticks= Temperaturas[:])
plt.grid()
plt.xlabel("Temperatura (K)")
plt.ylabel("Energia")
plt.legend()
plt.tight_layout()
```

```python
# Função linear para o ajuste
def linear_func(x, m, b):
    return m * x + b
```

```python
# Realiza o ajuste linear usando curve_fit
params, covariance = curve_fit(linear_func, np.array(temperaturas), np.array(energias))

# Obtém os parâmetros ajustados
m_fit, b_fit = params

# Criação do gráfico de dispersão e reta ajustada
plt.scatter(np.array(temperaturas), np.array(energias), label='Dados')
plt.plot(np.array(temperaturas), linear_func(np.array(temperaturas), m_fit, b_fit), 'r', label='Ajuste Linear: y = {:.2f}x + {:.2f}'.format(m_fit, b_fit))

plt.xlabel('Temperatura (K)')
plt.ylabel('Energia ')
plt.title('Ajuste Linear de Dados')
plt.legend()
plt.grid(True)
plt.xticks(Temperaturas)
plt.show()

print(m_fit, b_fit)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(temperaturas, volumes, label='Vol x Temp')
plt.xticks(ticks= Temperaturas[:])
plt.xlabel("Temperatura")
plt.ylabel("Volumes")
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig("Volume_X_Temperatura-NPT-BaO.png")
```
